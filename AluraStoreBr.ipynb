### Importa√ß√£o dos dados



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_1.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_4.csv"

loja = pd.read_csv(url)
loja2 = pd.read_csv(url2)
loja3 = pd.read_csv(url3)
loja4 = pd.read_csv(url4)

loja.head()
```

#1. An√°lise do faturamento


```python
# Calculando faturamento total (soma da coluna Pre√ßo)
faturamento1 = loja["Pre√ßo"].sum()
faturamento2 = loja2["Pre√ßo"].sum()
faturamento3 = loja3["Pre√ßo"].sum()
faturamento4 = loja4["Pre√ßo"].sum()

# Exibindo resultados
print("Faturamento Loja 1:", f"R$ {faturamento1:,.2f}".replace(",", "X").replace(".", ",").replace("X", "."))
print("Faturamento Loja 2:", f"R$ {faturamento2:,.2f}".replace(",", "X").replace(".", ",").replace("X", "."))
print("Faturamento Loja 3:", f"R$ {faturamento3:,.2f}".replace(",", "X").replace(".", ",").replace("X", "."))
print("Faturamento Loja 4:", f"R$ {faturamento4:,.2f}".replace(",", "X").replace(".", ",").replace("X", "."))
```

# 2. Vendas por Categoria


```python
# Fun√ß√£o para contar produtos por categoria
def produtos_por_categoria(df, nome_loja):
    resultado = df.groupby("Categoria do Produto")["Produto"].count().sort_values(ascending=False)
    print(f"\nüìä Categorias mais populares da {nome_loja}:")
    print(resultado)
    return resultado

# Aplicando para cada loja
categorias_loja1 = produtos_por_categoria(loja, "Loja 1")
categorias_loja2 = produtos_por_categoria(loja2, "Loja 2")
categorias_loja3 = produtos_por_categoria(loja3, "Loja 3")
categorias_loja4 = produtos_por_categoria(loja4, "Loja 4")
```

# 3. M√©dia de Avalia√ß√£o das Lojas

```python
# Calculando m√©dia das avalia√ß√µes
media_avaliacao1 = loja["Avalia√ß√£o da compra"].mean()
media_avaliacao2 = loja2["Avalia√ß√£o da compra"].mean()
media_avaliacao3 = loja3["Avalia√ß√£o da compra"].mean()
media_avaliacao4 = loja4["Avalia√ß√£o da compra"].mean()

# Exibindo resultados
print(f"M√©dia das avalia√ß√µes Loja 1: {media_avaliacao1:.2f}")
print(f"M√©dia das avalia√ß√µes Loja 2: {media_avaliacao2:.2f}")
print(f"M√©dia das avalia√ß√µes Loja 3: {media_avaliacao3:.2f}")
print(f"M√©dia das avalia√ß√µes Loja 4: {media_avaliacao4:.2f}")
```

# 4. Produtos Mais e Menos Vendidos

```python
# Fun√ß√£o para identificar produtos mais e menos vendidos
def analisar_produtos(df, nome_loja):
    contagem = df.groupby("Produto")["Produto"].count().sort_values(ascending=False)

    mais_vendido = contagem.idxmax()
    menos_vendido = contagem.idxmin()

    print(f"\nüè¨ {nome_loja}")
    print(f"‚û°Ô∏è Produto mais vendido: {mais_vendido} ({contagem.max()} vendas)")
    print(f"‚û°Ô∏è Produto menos vendido: {menos_vendido} ({contagem.min()} vendas)")

    # Mostrar tamb√©m o top 5 e bottom 5
    print("\nTop 5 produtos mais vendidos:")
    print(contagem.head(5))

    print("\nBottom 5 produtos menos vendidos:")
    print(contagem.tail(5))

# Aplicando para cada loja
analisar_produtos(loja, "Loja 1")
analisar_produtos(loja2, "Loja 2")
analisar_produtos(loja3, "Loja 3")
analisar_produtos(loja4, "Loja 4")
```

# 5. Frete M√©dio por Loja


```python
# Calculando custo m√©dio de frete
frete_medio1 = loja["Frete"].mean()
frete_medio2 = loja2["Frete"].mean()
frete_medio3 = loja3["Frete"].mean()
frete_medio4 = loja4["Frete"].mean()

# Exibindo resultados formatados em R$
print("Custo m√©dio de frete Loja 1:", f"R$ {frete_medio1:,.2f}".replace(",", "X").replace(".", ",").replace("X", "."))
print("Custo m√©dio de frete Loja 2:", f"R$ {frete_medio2:,.2f}".replace(",", "X").replace(".", ",").replace("X", "."))
print("Custo m√©dio de frete Loja 3:", f"R$ {frete_medio3:,.2f}".replace(",", "X").replace(".", ",").replace("X", "."))
print("Custo m√©dio de frete Loja 4:", f"R$ {frete_medio4:,.2f}".replace(",", "X").replace(".", ",").replace("X", "."))
```
